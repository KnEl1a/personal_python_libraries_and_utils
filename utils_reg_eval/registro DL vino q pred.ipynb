Es interesante notar que, despus de ajustar los hiperparmetros de tu modelo (aumentar la cantidad de unidades neuronales, el tamao del lote y el nmero de pocas), has observado una mejora en el \( R^2 \) pero un empeoramiento en el MAE. Esto puede parecer contradictorio a primera vista, pero hay razones subyacentes para este comportamiento.

### Explicacin del Comportamiento

#### 1. \( R^2 \) (Coeficiente de Determinacin)

El \( R^2 \) mide la proporcin de la varianza en la variable objetivo que es explicada por el modelo. Un aumento en \( R^2 \) indica que el modelo es capaz de capturar mejor la estructura de los datos y explicar ms de su variabilidad. Es posible que, al aumentar la complejidad del modelo (ms neuronas, ms capas), el modelo haya capturado patrones ms complejos en los datos, lo que se refleja en un mejor \( R^2 \).

#### 2. MAE (Mean Absolute Error)

El MAE es una mtrica que promedia los errores absolutos entre las predicciones y los valores reales. Si el MAE ha empeorado, significa que en promedio, los errores absolutos han aumentado. Esto puede suceder si el modelo ha comenzado a sobreajustarse (overfitting) o si ha ajustado mejor a ciertas partes de los datos mientras ha empeorado en otras.

### Razones Potenciales

#### Aumento en la Complejidad del Modelo

- **Captura de Patrones Complejos**: Al aumentar la cantidad de unidades neuronales y el nmero de pocas, el modelo puede capturar patrones ms complejos en los datos. Esto puede mejorar el \( R^2 \) porque el modelo explica mejor la varianza en los datos.
- **Ajuste a Ruido y Outliers**: Sin embargo, un modelo ms complejo tambin puede comenzar a ajustarse a ruidos o outliers en los datos. Esto puede aumentar el MAE porque los errores individuales pueden ser mayores.

#### Compromiso entre Bias y Varianza

- **Bias Reducido, Varianza Aumentada**: Aumentar la complejidad del modelo reduce el bias pero puede aumentar la varianza. El \( R^2 \) mejora porque el modelo ajusta mejor a los datos de entrenamiento, pero el MAE puede empeorar si el modelo empieza a sobreajustarse y comete errores ms grandes en algunas predicciones.

### Significado e Interpretacin

- **Mejora en \( R^2 \)**: Indica que el modelo est capturando ms de la variabilidad en los datos, lo que es una seal de que el modelo ha mejorado en su capacidad para explicar los datos.
- **Empeoramiento en MAE**: Indica que, aunque el modelo es mejor en trminos de varianza explicada, comete errores ms grandes en promedio. Esto puede ser un signo de sobreajuste (overfitting) o de sensibilidad a outliers.

### Estrategias para Mejorar

1. **Regularizacin**: Implementa tcnicas de regularizacin como L1, L2 o Dropout para evitar el sobreajuste.
2. **Validacin Cruzada**: Utiliza validacin cruzada para asegurarte de que el modelo generaliza bien y no est simplemente ajustndose a ruidos especficos del conjunto de entrenamiento.
3. **Anlisis de Outliers**: Revisa y maneja los outliers en tus datos, ya que pueden estar influyendo en el aumento del MAE.
4. **Ajuste de Hiperparmetros**: Contina ajustando los hiperparmetros, buscando un balance ptimo entre la complejidad del modelo y su capacidad de generalizacin.

En resumen, la mejora en \( R^2 \) y el empeoramiento en MAE sugiere que el modelo es ahora ms complejo y capaz de explicar ms variabilidad en los datos, pero tambin est cometiendo errores ms grandes en algunas predicciones, posiblemente debido a sobreajuste o a la presencia de outliers.
