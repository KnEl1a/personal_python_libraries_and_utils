Es interesante notar que, después de ajustar los hiperparámetros de tu modelo (aumentar la cantidad de unidades neuronales, el tamaño del lote y el número de épocas), has observado una mejora en el \( R^2 \) pero un empeoramiento en el MAE. Esto puede parecer contradictorio a primera vista, pero hay razones subyacentes para este comportamiento.

### Explicación del Comportamiento

#### 1. \( R^2 \) (Coeficiente de Determinación)

El \( R^2 \) mide la proporción de la varianza en la variable objetivo que es explicada por el modelo. Un aumento en \( R^2 \) indica que el modelo es capaz de capturar mejor la estructura de los datos y explicar más de su variabilidad. Es posible que, al aumentar la complejidad del modelo (más neuronas, más capas), el modelo haya capturado patrones más complejos en los datos, lo que se refleja en un mejor \( R^2 \).

#### 2. MAE (Mean Absolute Error)

El MAE es una métrica que promedia los errores absolutos entre las predicciones y los valores reales. Si el MAE ha empeorado, significa que en promedio, los errores absolutos han aumentado. Esto puede suceder si el modelo ha comenzado a sobreajustarse (overfitting) o si ha ajustado mejor a ciertas partes de los datos mientras ha empeorado en otras.

### Razones Potenciales

#### Aumento en la Complejidad del Modelo

- **Captura de Patrones Complejos**: Al aumentar la cantidad de unidades neuronales y el número de épocas, el modelo puede capturar patrones más complejos en los datos. Esto puede mejorar el \( R^2 \) porque el modelo explica mejor la varianza en los datos.
- **Ajuste a Ruido y Outliers**: Sin embargo, un modelo más complejo también puede comenzar a ajustarse a ruidos o outliers en los datos. Esto puede aumentar el MAE porque los errores individuales pueden ser mayores.

#### Compromiso entre Bias y Varianza

- **Bias Reducido, Varianza Aumentada**: Aumentar la complejidad del modelo reduce el bias pero puede aumentar la varianza. El \( R^2 \) mejora porque el modelo ajusta mejor a los datos de entrenamiento, pero el MAE puede empeorar si el modelo empieza a sobreajustarse y comete errores más grandes en algunas predicciones.

### Significado e Interpretación

- **Mejora en \( R^2 \)**: Indica que el modelo está capturando más de la variabilidad en los datos, lo que es una señal de que el modelo ha mejorado en su capacidad para explicar los datos.
- **Empeoramiento en MAE**: Indica que, aunque el modelo es mejor en términos de varianza explicada, comete errores más grandes en promedio. Esto puede ser un signo de sobreajuste (overfitting) o de sensibilidad a outliers.

### Estrategias para Mejorar

1. **Regularización**: Implementa técnicas de regularización como L1, L2 o Dropout para evitar el sobreajuste.
2. **Validación Cruzada**: Utiliza validación cruzada para asegurarte de que el modelo generaliza bien y no está simplemente ajustándose a ruidos específicos del conjunto de entrenamiento.
3. **Análisis de Outliers**: Revisa y maneja los outliers en tus datos, ya que pueden estar influyendo en el aumento del MAE.
4. **Ajuste de Hiperparámetros**: Continúa ajustando los hiperparámetros, buscando un balance óptimo entre la complejidad del modelo y su capacidad de generalización.

En resumen, la mejora en \( R^2 \) y el empeoramiento en MAE sugiere que el modelo es ahora más complejo y capaz de explicar más variabilidad en los datos, pero también está cometiendo errores más grandes en algunas predicciones, posiblemente debido a sobreajuste o a la presencia de outliers.
